## **III. 📌 Extracción (E - Extract)**

La fase de **Extracción** constituye el punto de partida del flujo de trabajo **ETL**. En esta etapa, el objetivo primordial es establecer una conexión segura y eficiente con la fuente de origen para obtener los **datos crudos** de Telecom X.

El éxito de esta fase es crítico, ya que cualquier error u omisión en la captura de datos afectaría la validez de los análisis de **fuga de clientes** y las conclusiones estratégicas posteriores.

### **3.1. Extracción de Datos de la API**

Para este análisis, consumiremos un endpoint que entrega la información en formato **JSON**. Este método garantiza que estamos trabajando con la versión más reciente de la base de datos de clientes, la cual incluye las dimensiones demográficas, de servicios y de facturación necesarias para el estudio de **Churn**.

```python
# =============================================================================
# 3.1. EXTRACCIÓN DE DATOS DESDE LA API (URL VALIDADA)
# =============================================================================

# URL en formato raw para permitir la descarga directa del contenido JSON
url = "https://raw.githubusercontent.com/ingridcristh/challenge2-data-science-LATAM/main/TelecomX_Data.json"

try:
    # Realizar la petición GET
    response = requests.get(url)

    # Verificar éxito de la conexión
    if response.status_code == 200:
        data_raw = response.json()
        print(f"Éxito: Se han extraído {len(data_raw)} registros correctamente.")

        # Muestra el primer registro para inspección inicial
        print("\nEjemplo de estructura del primer registro:")
        print(data_raw[0])
    else:
        print(f"Error en la conexión: Código de estado {response.status_code}")

except Exception as e:
    print(f"Error inesperado al intentar extraer los datos: {e}")
```

### **3.2. Estructuración y Normalización**

Dado que el archivo original tiene una estructura jerárquica (formato JSON con diccionarios anidados), es necesario realizar un proceso de **aplanamiento** o normalización.

En este paso, transformaremos los objetos complejos como `customer`, `phone`, `internet` y `account` en columnas individuales. Esto nos permitirá consolidar un **DataFrame** estructurado donde cada fila represente un cliente único y cada columna una de sus características, facilitando así la manipulación de datos con **Pandas**.

####**3.2.1. Aplanamiento de Datos (Flattening)**

El primer paso es utilizar json_normalize. Esto toma los diccionarios anidados dentro del JSON y los convierte en columnas individuales, eliminando la jerarquía compleja.

```python
# =============================================================================
# 3.2.1. NORMALIZACIÓN DE LA ESTRUCTURA JSON
# =============================================================================

# Expandimos los diccionarios anidados en columnas independientes
df_clientes = pd.json_normalize(data_raw)

# Validamos la nueva forma del conjunto de datos
print(f"Dataset normalizado con {df_clientes.shape[1]} columnas.")
```

####**3.2.2. Inspección de Columnas Generadas**

Una vez normalizado, es vital revisar los nombres de las columnas. La función suele crear nombres con puntos (ej. cliente.nombre) para representar la ruta original del JSON.

```python
# =============================================================================
# 3.2.2. LISTADO DE VARIABLES NORMALIZADAS
# =============================================================================

# Listamos todas las columnas para verificar la ruta de normalización
print("Variables disponibles tras el aplanamiento:")
print(df_clientes.columns.tolist())
```

Es un paso crítico de normalización; sin él, el sistema podría interpretar `"Fiber optic"` y `" Fiber optic"` como dos servicios distintos, lo que sesgaría gravemente los resultados del análisis.

### **4.3. Conversión de Tipos de Datos (Casting)**

Este proceso técnico consiste en la **asignación correcta del tipo de dato** para cada columna, asegurando que Python interprete la información según su naturaleza lógica. Tras la importación desde la API, variables financieras críticas como los cargos totales y mensuales se cargan como objetos de texto (strings) debido a formatos de origen o presencia de caracteres especiales.

Realizaremos el **casting** o transformación forzada de estas variables a formatos de **punto flotante (float)**. Esta conversión es el filtro final de calidad: nos permite detectar errores de formato que pasaron inadvertidos y prepara el terreno para la resolución técnica de los registros con antigüedad cero, garantizando que el dataset posea la precisión necesaria para las fases analíticas de **Telecom X**. Sin esta etapa, sería imposible calcular promedios o generar visualizaciones estadísticas confiables.

#### **4.3.1. Identificación y Mapeo de Columnas Financieras**

Tras la inspección inicial en la Fase III, se identificaron variables que contienen información económica pero que están almacenadas como tipo `object` (texto). Para realizar el análisis estadístico y de correlación, es imperativo mapearlas y prepararlas para su conversión a formato numérico:

* **`account.Charges.Monthly`**: Representa el cargo mensual del cliente.
* **`account.Charges.Total`**: Representa el acumulado histórico, la cual presenta espacios en blanco que deben ser tratados como nulos durante la conversión.

**Objetivo:** Asegurar que estas columnas permitan operaciones aritméticas en las fases posteriores.correlación.

```python
# =============================================================================
# 4.3.1. VERIFICACIÓN PREVIA AL CASTING
# =============================================================================

# Listamos las columnas identificadas para el mapeo
cols_a_mapear = ['account.Charges.Monthly', 'account.Charges.Total']

# Mostramos el tipo de dato actual y una muestra para detectar espacios en blanco
print("--- ESTADO PREVIO DE COLUMNAS FINANCIERAS ---")
print(df_consolidado[cols_a_mapear].dtypes)
print("\nMuestra de valores (primeros 5):")
print(df_consolidado[cols_a_mapear].head())
```

#### **4.3.2. Conversión de Variables Financieras a Flotantes**

Ejecutamos la conversión técnica a tipo `float`. Para la columna `account.Charges.Total`, implementamos el parámetro `errors='coerce'`.

Esta es una decisión de diseño fundamental: al encontrar los registros inconsistentes o con espacios en blanco detectados en la auditoría de la **Sección 3.4**, el sistema los transformará automáticamente en `NaN` (nulos técnicos). Esto garantiza que el cuaderno no se detenga por errores de tipo y permite un tratamiento controlado en la siguiente subsección (4.3.3).

```python
# =============================================================================
# 4.3.2. CONVERSIÓN DE VARIABLES FINANCIERAS A FLOTANTES
# =============================================================================

# Definimos las columnas identificadas en la sección 4.3.1
cols_financieras = ['account.Charges.Monthly', 'account.Charges.Total']

# Aplicamos la conversión forzando errores (como espacios vacíos) a NaN
for col in cols_financieras:
    df_consolidado[col] = pd.to_numeric(df_consolidado[col], errors='coerce')

print("--- REPORTE DE CONVERSIÓN (CASTING) ---")
print(df_consolidado[cols_financieras].dtypes)

# Validamos cuántos nulos reales tenemos ahora para preparar la sección 4.3.3
nulos_post_casting = df_consolidado['account.Charges.Total'].isna().sum()
print(f"\n✅ Conversión exitosa. Nulos detectados en TotalCharges: {nulos_post_casting}")
```

#### **4.3.3. Tratamiento de Valores Nulos por Conversión (Imputación Técnica)**

Tras el proceso de casting, los registros que presentaban inconsistencias (específicamente aquellos con antigüedad cero que no habían completado su primer ciclo de facturación) han sido transformados en valores `NaN`.

En lugar de eliminar estos registros, aplicamos una **imputación técnica de 0.0**. Esta decisión se basa en la lógica de negocio: un cliente nuevo (`tenure` = 0) legítimamente tiene un cargo total acumulado de cero. De esta forma, preservamos la integridad de la muestra sin introducir sesgos.

```python
# =============================================================================
# 4.3.3. IMPUTACIÓN DE VALORES NULOS POST-CONVERSIÓN
# =============================================================================

# Identificamos los nulos antes de la corrección
nulos_antes = df_consolidado['account.Charges.Total'].isna().sum()

# Aplicamos la imputación con 0.0 a los valores nulos resultantes del casting
df_consolidado['account.Charges.Total'] = df_consolidado['account.Charges.Total'].fillna(0.0)

# Verificación final de la columna
nulos_despues = df_consolidado['account.Charges.Total'].isna().sum()

print(f"✅ Valores nulos tratados: {nulos_antes}")
print(f"📊 Nulos residuales en TotalCharges: {nulos_despues}")
print(f"📈 Tipo de dato final: {df_consolidado['account.Charges.Total'].dtype}")
```

### **4.4. Validación de Consistencia y Corrección de Errores**

Una vez que el dataset es numéricamente completo, procedemos a una auditoría de coherencia lógica. En esta fase, validamos que los valores financieros guarden una relación matemática razonable entre sí.

Detectar discrepancias donde los cargos totales no coinciden con el producto de la mensualidad y la antigüedad nos permite identificar errores de arrastre de la fuente original o casos atípicos (outliers) que podrían distorsionar los modelos de predicción de fuga. Es el último filtro de seguridad antes de proceder con la generación de nuevas métricas.

#### **4.4.1. Validación Específica de Clientes con Tenencia Cero**

Para los clientes identificados con una antigüedad (`tenure`) de 0 meses, es imperativo que sus cargos totales sean consistentes con su estado de "recién contratados".

Tras la imputación realizada en la sección anterior, en esta celda validamos que no existan remanentes de inconsistencias donde un cliente nuevo presente acumulados históricos, asegurando que la relación entre antigüedad y facturación sea lógicamente impecable.

```python
# =============================================================================
# 4.4.1. VALIDACIÓN DE CONSISTENCIA EN CLIENTES NUEVOS
# =============================================================================

# Filtramos clientes con tenure 0 para verificar sus cargos
clientes_nuevos = df_consolidado[df_consolidado['customer.tenure'] == 0]

# Verificamos si después de la sección 4.3.3 todos tienen TotalCharges en 0
inconsistencias_restantes = clientes_nuevos[clientes_nuevos['account.Charges.Total'] != 0].shape[0]

print(f"--- VALIDACIÓN DE COHERENCIA (TENURE 0) ---")
print(f"👤 Total de clientes con antigüedad cero: {len(clientes_nuevos)}")
print(f"✅ Clientes con cargos consistentes (Total = 0): {len(clientes_nuevos) - inconsistencias_restantes}")

if inconsistencias_restantes > 0:
    print(f"⚠️ Corrigiendo {inconsistencias_restantes} registros para alinearlos a la lógica de negocio...")
    df_consolidado.loc[df_consolidado['customer.tenure'] == 0, 'account.Charges.Total'] = 0.0
    print("✅ Ajuste de consistencia finalizado.")
else:
    print("✅ No se detectaron inconsistencias lógicas en los cargos de clientes nuevos.")
```

#### **4.4.2. Validación de Consistencia en TotalCharges**

Realizamos un chequeo cruzado para verificar que el cargo total acumulado sea razonable en función del cargo mensual y la antigüedad del cliente. Aunque no esperamos una igualdad exacta debido a variaciones en servicios opcionales, esta validación permite detectar errores graves de carga en la base de datos original.

```python
# =============================================================================
# 4.4.2. AUDITORÍA DE RAZONABILIDAD FINANCIERA
# =============================================================================

# Creamos una columna temporal para calcular la diferencia teórica
# (Cargo Mensual * Antigüedad) vs Cargo Total
df_consolidado['Check_Financiero'] = df_consolidado['account.Charges.Monthly'] * df_consolidado['customer.tenure']

# Identificamos casos donde el Cargo Total es drásticamente mayor al teórico (indicador de error)
# Usamos un margen de tolerancia para variaciones normales
casos_atipicos = df_consolidado[df_consolidado['account.Charges.Total'] > (df_consolidado['Check_Financiero'] * 1.5)].shape[0]

print(f"--- REPORTE DE RAZONABILIDAD ---")
print(f"🔍 Registros con desviaciones financieras atípicas: {casos_atipicos}")

# Eliminamos la columna temporal para mantener el dataset limpio
df_consolidado.drop(columns=['Check_Financiero'], inplace=True)

print("✅ Auditoría de consistencia financiera finalizada.")
```

### **4.5 Ingeniería de Datos**

En esta etapa estratégica, trascendemos la limpieza básica para enfocarnos en la **generación de valor añadido** a través de la creación de nuevas variables. El propósito es sintetizar comportamientos complejos en métricas sencillas que faciliten la detección de **patrones de abandono**. Al derivar nuevas columnas, enriquecemos el dataset original y preparamos el terreno para un análisis mucho más sofisticado, permitiendo que las variables demográficas y financieras interactúen para revelar **segmentos de riesgo** que antes eran invisibles en los datos primarios.

#### **4.5.1. Creación de la Columna "Cuentas_Diarias"**

Esta variable permite normalizar el gasto de los clientes a una escala diaria, convirtiéndose en una métrica de intensidad de consumo. Ayuda a identificar si los clientes con mayor frecuencia de gasto diario presentan una mayor propensión al abandono (*Churn*).

El cálculo se realiza dividiendo el cargo mensual entre un estándar comercial de 30 días:

$$Cuentas\_Diarias = \frac{account.Charges.Monthly}{30}$$

```python
# =============================================================================
# 4.5.1. CREACIÓN DE LA COLUMNA "CUENTAS_DIARIAS"
# =============================================================================

# Cálculo del gasto diario promedio
df_consolidado['Cuentas_Diarias'] = df_consolidado['account.Charges.Monthly'] / 30

# Verificación de la nueva métrica
print("--- REPORTE DE INGENIERÍA DE DATOS (4.5.1) ---")
print(f"✅ Nueva variable 'Cuentas_Diarias' creada.")
print(f"📈 Valor promedio de gasto diario: {df_consolidado['Cuentas_Diarias'].mean():.2f}")
print("\nMuestra de resultados:")
print(df_consolidado[['account.Charges.Monthly', 'Cuentas_Diarias']].head())
```

#### **4.5.2. Segmentación de Antigüedad (Tenure)**

Transformamos la variable numérica de meses de permanencia en una variable categórica de rangos temporales. El objetivo es identificar el "periodo crítico" en el ciclo de vida del cliente mediante la siguiente segmentación:

$$Segmento\_Tenure = \begin{cases} Rookie & 0 \le x \le 12 \\ Estable & 13 \le x \le 24 \\ Fidelizado & 25 \le x \le 48 \\ Senior & x > 48 \end{cases}$$

Esta categorización permitirá a **Telecom X** diseñar estrategias de retención diferenciadas según la lealtad del usuario.

```python
# =============================================================================
# 4.5.2. SEGMENTACIÓN DE ANTIGÜEDAD (TENURE)
# =============================================================================

# Definición de límites y etiquetas para los segmentos
bins = [0, 12, 24, 48, float('inf')]
labels = ['Rookie', 'Estable', 'Fidelizado', 'Senior']

# Aplicación de la segmentación (incluyendo el límite inferior para clientes nuevos)
df_consolidado['Segmento_Tenure'] = pd.cut(df_consolidado['customer.tenure'],
                                           bins=bins,
                                           labels=labels,
                                           include_lowest=True)

print("--- REPORTE DE SEGMENTACIÓN (4.5.2) ---")
print("✅ Variable 'Segmento_Tenure' generada.")
print("\nDistribución por Segmento:")
print(df_consolidado['Segmento_Tenure'].value_counts().sort_index())
```

#### **5.1.1.2. Gráfico de Sectores (Proporción Porcentual)**

Complementamos el análisis univariado de la variable objetivo mediante una visualización de **composición relativa**. Mientras que el gráfico de barras anterior nos permitió observar frecuencias absolutas, el gráfico de sectores (pie chart) facilita la comprensión de la **cuota de participación** de la evasión dentro de la base total de clientes.

Para esta visualización, implementamos un realce visual dinámico (**Explode**):
* **Enfoque Estratégico:** Se separa ligeramente la porción correspondiente a los clientes que abandonaron la compañía (`Churn = Yes`).
* **Objetivo:** Dirigir la atención del analista hacia el segmento crítico, permitiendo una interpretación inmediata de la magnitud del problema de retención en **Telecom X**.

```python
# =============================================================================
# 5.1.1.2. GRÁFICO DE SECTORES (PROPORCIÓN PORCENTUAL)
# =============================================================================

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6, 6))

# Cálculo de la distribución para el gráfico circular
churn_counts = df_consolidado['Churn'].value_counts()

# Crear una lista de ceros para explode del mismo tamaño que churn_counts
explode_values = [0] * len(churn_counts)

# Identificar el índice de 'Yes' y aplicar el 'explode' deseado
# Esto asume que 'Yes' estará presente, lo cual es fundamental para el análisis de Churn
if 'Yes' in churn_counts.index:
    yes_index = churn_counts.index.get_loc('Yes')
    explode_values[yes_index] = 0.05 # Resaltamos ligeramente la categoría de interés (Churn 'Yes')

# Generación de la visualización con énfasis en porcentajes
plt.pie(churn_counts,
        labels=churn_counts.index,
        autopct='%1.1f%%',
        startangle=140,
        colors=sns.color_palette('viridis', len(churn_counts)),
        explode=explode_values) # Usamos la lista de explode generada dinámicamente

plt.title('Participación Porcentual: Evasión vs. Permanencia')
plt.show()
```

El gráfico de sectores, también conocido como "pie chart", complementa el análisis de frecuencia al mostrar la **participación porcentual** de la evasión y la permanencia de clientes. Para **Telecom X**, esta visualización es fundamental porque permite comprender rápidamente el **peso relativo** de cada estado de Churn. Al resaltar visualmente la proporción (26.54%) de clientes que se han fugado frente a los que permanecen (73.46%), este gráfico enfatiza el **grado de desbalanceo** del dataset. Este conocimiento es crucial para el diseño de modelos predictivos, ya que la clase minoritaria (Churn 'Yes') es a menudo la de mayor interés para la empresa, y su representación en el dataset influirá en la elección de las técnicas de modelado para evitar sesgos.

```python
# =============================================================================
# 5.2.1. VARIABLES CATEGÓRICAS VS. EVASIÓN (CRUCE PROPORCIONAL)
# =============================================================================

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Seleccionamos las variables clave identificadas en el análisis univariado
vars_to_cross = [
    'account.Contract',
    'internet.InternetService',
    'account.PaymentMethod',
    'account.PaperlessBilling'
]

# Configuración de la cuadrícula (2 filas, 2 columnas)
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(16, 14))
axes = axes.flatten()

for i, col in enumerate(vars_to_cross):
    # Creamos una tabla de contingencia normalizada (proporciones)
    cross_tab = pd.crosstab(df_consolidado[col], df_consolidado['Churn'], normalize='index')

    # Graficamos las barras apiladas
    cross_tab.plot(kind='bar', stacked=True, ax=axes[i], color=['#2a9d8f', '#e76f51'], alpha=0.8)

    axes[i].set_title(f'Impacto de {col} en el Churn', fontsize=12, fontweight='bold')
    axes[i].set_ylabel('Proporción de Clientes')
    axes[i].set_xlabel('')
    axes[i].legend(title='Churn', loc='upper right')
    axes[i].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

####**5.3.1. Matriz de Correlación**

En este punto técnico, generaremos una **representación visual (Heatmap)** que condensa los coeficientes de correlación de todas las variables numéricas del estudio. El objetivo es identificar **dependencias estadísticas** claras; por ejemplo, una correlación cercana a **+1** nos indicará que las variables crecen juntas (como la antigüedad y los cargos totales), mientras que valores cercanos a **0** sugerirán que no existe una relación lineal entre ellas.

Esta matriz es la herramienta definitiva para detectar **multicolinealidad** y confirmar qué factores financieros tienen el peso matemático más significativo en la estructura de datos de **Telecom X**.

```python
# =============================================================================
# 5.3.1. MATRIZ DE CORRELACIÓN (HEATMAP)
# =============================================================================

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Seleccionamos solo las variables numéricas para el cálculo de correlación
df_numerico = df_consolidado.select_dtypes(include=[np.number])

# Calculamos la matriz de correlación de Pearson
corr_matrix = df_numerico.corr()

# Configuración del gráfico
plt.figure(figsize=(12, 8))

# Generamos el Mapa de Calor (Heatmap)
sns.heatmap(corr_matrix,
            annot=True,          # Mostramos los valores numéricos
            fmt=".2f",           # Formato de dos decimales
            cmap='coolwarm',     # Escala de colores (azul frío a rojo cálido)
            linewidths=0.5,
            square=True)

plt.title('Matriz de Correlación de Variables Numéricas', fontsize=15, fontweight='bold')
plt.show()
```

La **Matriz de Correlación (Heatmap)** de la sección **5.3.1.** nos proporciona una visión visual de las relaciones lineales entre las variables numéricas. Los valores cercanos a +1 indican una fuerte correlación positiva, mientras que los valores cercanos a -1 indican una fuerte correlación negativa, y los cercanos a 0 sugieren poca o ninguna relación lineal.

Para **Telecom X**, este heatmap revela lo siguiente:

*   **`customer.tenure` (Antigüedad) y `account.Charges.Total` (Cargos Totales)**: Muestran una **correlación muy fuerte y positiva (0.83)**. Esto es esperado e intuitivo, ya que cuanto más tiempo permanece un cliente, más cargos totales acumula. Este valor resalta el impacto de la antigüedad en el valor de vida del cliente (LTV).
